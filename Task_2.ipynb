```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import datetime
```

```python
df=pd.read_csv('/content/superstore_final_dataset (1).csv', encoding='latin')
```

**Converting Order_Date & Ship_Date column from object to datetime datatype.**

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], dayfirst=True)
df['Ship_Date']=pd.to_datetime(df['Ship_Date'],dayfirst=True)
```

```python
df.info()
```

```python
# Check for null values
df.isnull().sum()
```

```python
# Dropping null values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#Descriptive statistics
df.describe()
```

```python
# Qualitative characters
df.describe(include='O')
```

```python
df = df[df['Sales'] > 0]
```

```python
#Average  of Sales
average_sales=df['Sales'].mean()
print(average_sales)
```

```python
# Summary of sales distribution
sales_distribution=df['Sales'].describe()
print("\nSales Distribution:\n",sales_distribution)
```

**Revenue**

```python
#Revenue
Revenue=df['Sales'].sum()
print('Revenue:',Revenue)
```

```python
df['order_Year']=df['Order_Date'].dt.year
df['order_quater']=df['Order_Date'].dt.quarter
df['Order_Month']=df['Order_Date'].dt.month

print("DataFrame with new time components:")
df.head()
```

```python
yearly_sales = df.groupby('order_Year')['Sales'].sum()
print("\nYearly Sales:\n")
yearly_sales.head()
```

```python
quarterly_sales = df.groupby(['order_Year', 'order_quater'])['Sales'].sum()
print("\nQuarterly Sales:\n")
quarterly_sales.head()
```

```python
Monthly_sales=df.groupby(['order_Year','Order_Month'])['Sales'].sum()
print('\nMonthly_sales\n')
display(Monthly_sales.head())
```

Profit margin cannot be calculated because the dataset does not include cost or profit information. Without these attributes, profitability cannot be derived from sales data alone, so profit-based KPIs were excluded to avoid misleading results.

## Avarage order value (AOV)

```python
aov=df.groupby('Order_ID')['Sales'].sum().mean()
print('AOV:',aov)
```

# Visualization

## Revenue Trend Over Time

```python
import matplotlib.pyplot as plt
# Revenue Trend Over Time
df['Month'] = df['Order_Date'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,6))
monthly_sales.plot()
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

**Conclusion** :
The monthly revenue trend shows a clear upward growth pattern over time, indicating that overall sales performance is improving year by year. While revenue fluctuates significantly from month to month, the peaks become higher in later years, suggesting successful expansion and increasing market demand.

## Top products by Revenue

```python
top_products=df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure()
top_products.plot(kind='bar')
plt.title("Top 10 Products by Revenue")
plt.ylabel("Revenue")
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Conclusion** : The bar chart highlights a highly skewed revenue distribution, where a small number of products contribute a large share of total sales. This indicates strong product concentration within the portfolio.

# Regional Performance

```python
regional_sales = df.groupby('Region')['Sales'].sum()

plt.figure()
regional_sales.plot(kind='bar')
plt.title("Revenue by Region")
plt.ylabel("Revenue")
plt.show()
```

**Conclusion**  : The regional revenue analysis reveals significant variation in sales performance across regions, indicating uneven market penetration and growth opportunities.
